```python
from os import stat
import matplotlib.pyplot as plt
```

```python
csv_file = 'pokemon_stats.csv'
with open(csv_file) as f:
    csv_list = [[val.strip() for val in r.split(",")] for r in f.readlines()]

(_, *header), *data = csv_list
poke_dict = {}
for row in data:
    key, *values = row   
    poke_dict[key] = {key: value for key, value in zip(header, values)}
```

```python
def generation_frequencys(dict):
    arr = {}
    for p in dict:
            if dict.get(p).get('generation') in arr:
                arr[dict.get(p).get('generation')] += 1
            else:
                arr.update({dict.get(p).get('generation'): 1})
    stats_frequency.update({'generation': arr})
    return stats_frequency

generation_frequencys(poke_dict)
```

```python
height_m_frequency={
    character: value / sum(stats_frequency['height_m'].values())
    for character, value in stats_frequency['height_m'].items()
}
height_m_frequency
```

```python
def calculate_pokemon_commonality(pokemon):
    score = gen_frequency[poke_dict[pokemon]['generation']] + type1_frequency[poke_dict[pokemon]['type1']] + type2_frequency[poke_dict[pokemon]['type2']] + weight_kg_frequency[poke_dict[pokemon]['weight_kg']] + height_m_frequency[poke_dict[pokemon]['height_m']]
    return score 
calculate_pokemon_commonality('Bulbasaur')
```

```python
import operator

def sort_by_pokemon_commonality(pokemons):
    sort_by = operator.itemgetter(1)
    return sorted(
        [(pokemon, calculate_pokemon_commonality(pokemon)) for pokemon in pokemons],
        key=sort_by,
        reverse=True,
    )

def display_word_table(pokemon_commonalities):
    for (pokemon, freq) in pokemon_commonalities:
        print(f"{pokemon:<10} | {freq:<5.2}")
```

```python
sort_by_pokemon_commonality(poke_dict.keys())
```

```python
display_word_table(sort_by_pokemon_commonality(poke_dict.keys()))
```
